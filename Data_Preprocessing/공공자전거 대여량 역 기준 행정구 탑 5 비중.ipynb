```python
import numpy as np
def extract_station(row) : 
    row = row.copy()
    station = row.get('name',"0. drop").split()
    if len(station) > 1 :
        if station[1].endswith('역') :
            row['station'] = station[1]
        else :
            row['station'] = np.nan
    else :
        row['station'] = np.nan
    return row
```

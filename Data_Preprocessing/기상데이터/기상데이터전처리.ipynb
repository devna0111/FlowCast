# 기상데이터 전처리

```python
import pandas as pd
```

```python
gu = ['강남구', '강동구', '강북구', '강서구', '관악구', '광진구', '구로구', '금천구', '노원구', '도봉구','동대문구', '동작구', '마포구', '서대문구', '서초구', '성동구', '성북구', '송파구', '양천구', '영등포구', '용산구', '은평구', '종로구', '중구', '중랑구']
```

```python
df = pd.DataFrame(columns=['연도','월','날','시간','강수','기온','습도','풍속','행정구'])
for data in gu :
    file_name = f'C:/Users/Admin/OneDrive/Desktop/21~25 지역구 기상정리/21~25 지역구 기상정리/{data}_일기_2021_2025.csv'
    df_gu = pd.read_csv(file_name)
    df_gu['행정구']=data
    df = pd.concat([df,df_gu])
df
```

```python
df['기온'] = round(df['기온'].astype('float64'),1)
```

```python
df.isna().any()
```

```python
df['연도'] = df['연도'].astype('int32')
df['월'] = df['월'].astype('int32')
df['날'] = df['날'].astype('int32')
df['시간'] = df['시간'].astype('int32')
df
```

```python
df.to_csv('연도월일시간온강수기온습도풍속행정구.csv',index=False)
```
